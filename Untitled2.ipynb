```python
import re
```

```python
def searchSSN(text):
    q = [m.group() for m in re.finditer(r'\d\d\d-\d\d-\d\d\d\d', text)]
    return q
        
```

```python
def searchPESEL(text):
    q = [m.group() for m in re.finditer(r'\d\d\d\d\d\d\d\d\d\d\d', text)]
    return q
```

```python
def searchCARD(text):
    q = [m.group() for m in re.finditer(r'(\d\d\d\d \d\d\d\d \d\d\d\d \d\d\d\d|\d\d\d\d\d\d\d\d\d\d\d\d\d\d\d\d)', text)]
    return q
```

```python
def extractEmails(text):
    q = [m.group() for m in re.finditer(r'[a-zA-Z0-9._-]+@[a-zA-Z0-9._-]+\.[a-zA-Z0-9_-]+', text)]
    return q
```

```python
def extractIPs(text):
    q = [m.group() for m in re.finditer(r'\d+\.\d+\.\d+\.\d+', text)]
    return q
```

```python
def findLoginPassword(text):
    slownik ={
        'password' : (text.find('password') != -1) or (text.lower().find('pws') != -1),
        'login' : text.find('login') != -1,
        'username' : text.find('user') != -1,
        #'scram' : re.search('\:\/\/\w+\:', text) != -1,
    }
    return slownik
```

```python
def digits(text):
    return all(map(lambda x: x.isdigit(), text.split()))
```

```python
start = time.time()
for ii,d in enumerate(db.dane.find()):
    #print(d['lang'])
    #print(d['langProb'])
    if d['langProb'][0][1] > 0.7:
        #print('\t', d['langProb'][0][0])
        db.dane.update({'_id':d['_id']},{'$set':{'lang':d['langProb'][0][0]}})
    else:
        #print('\t', '__label__not_a_language')
        db.dane.update({'_id':d['_id']},{'$set':{'lang':'__label__not_a_language'}})
    #if d['lang'] == '__label__en':
    #    print(d['payload'])
    if ii % 10000 == 0:
        print(ii, time.time() - start)
print("All done ...", time.time() - start)
```

```python
start = time.time()
for ii, d in enumerate(db.dane.find()):
    update = {}
    if len(extractIPs(d['payload'])) > 0:
        update['ip'] = True
        update['ips'] = extractIPs(d['payload'])[:10]
    if len(extractEmails(d['payload'])) > 0:
        update['e-mail'] = True
        update['e-mails'] = extractEmails(d['payload'])[:10]
    if searchSSN(d['payload']):
        update['ssn'] = True
    sec = findLoginPassword(d['payload'])
    if sec['login']:
        update['login'] = True
    if sec['password']:
        update['password'] = True
    if sec['username']:
        update['username'] = True
    if update != {}:
        pass
        #print(update)
        db.dane.update_one({'_id':d['_id']},{'$set':update})
    if ii % 10000 == 0:
        print(ii, time.time() - start)
print("All done ...", time.time() - start)
```
