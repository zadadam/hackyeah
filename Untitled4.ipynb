```python
client = pymongo.MongoClient(port=27017)
db=client.exatel
```

```python
import numpy as np
import torch 
```

```python
from flair.embeddings import BytePairEmbeddings
from flair.embeddings import DocumentPoolEmbeddings, Sentence
embedding = BytePairEmbeddings('multi')
doc = DocumentPoolEmbeddings([embedding])
```

```python
start = time.time()
errors =[]
for ii,d in enumerate(db.dane.find()):
    if ii < 37694:
        continue
    if d['lang'] != '__label__not_a_language':
        try:
            sentence1 = Sentence(' '.join(d['payload'].split()))
            doc.embed(sentence1)
            vec = sentence1.get_embedding()
            norm = torch.sqrt(torch.dot(vec,vec))
            vecNorm = vec / norm
            db.dane.update_one({'_id':d['_id']}, {'$set':{'vec':(vecNorm).tolist()}})
        except:
            errors.append((ii,d['_id']))
    if ii % 1000 == 0:
        print(ii, time.time() - start)
print(time.time() - start)
```
